```python
# üìä Simple Sales Data Analysis in Google Colab (Encoding Error Fixed)

# Step 1: Import Libraries

import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Upload your CSV file

from google.colab import files
uploaded = files.upload()  # Upload 'sales_data_sample.csv'

# Step 3: Load Dataset with encoding handling

try:
    df = pd.read_csv('sales_data_sample.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('sales_data_sample.csv', encoding='latin1')

# Step 4: Preview the data

print("‚úÖ File loaded successfully!\n")
print("First 5 rows of dataset:")
print(df.head())

# Step 5: Check column names

print("\nColumn names:")
print(df.columns)

# Step 6: Adjust column names (change if needed)

# For the sample sales dataset, typical columns are: 'PRODUCTLINE', 'SALES'

product_col = 'PRODUCTLINE'
sales_col = 'SALES'

# Step 7: Calculate Total Sales

total_sales = df[sales_col].sum()
print(f"\nüí∞ Total Sales: {total_sales:,.2f}")

# Step 8: Find Best-Selling Product

best_selling = df.groupby(product_col)[sales_col].sum().idxmax()
best_sales_value = df.groupby(product_col)[sales_col].sum().max()
print(f"\nüèÜ Best-Selling Product: {best_selling} (${best_sales_value:,.2f})")

# Step 9: Create Basic Sales Report

sales_by_product = df.groupby(product_col)[sales_col].sum().sort_values(ascending=False)
print("\nüìà Sales Report by Product:")
print(sales_by_product)

# Step 10: Plot the results

plt.figure(figsize=(10,6))
sales_by_product.plot(kind='bar', color='cornflowerblue')
plt.title('Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
